- `date`: The date of the data in yyyy-mm-dd format. The data is correct as of 2359 hours on that date.
- `state`: The name of the state or federal territory in Malaysia where the PKRC is located.
- `beds`: The total number of beds available in the PKRC as of the date.
- `admitted_pui`: The number of persons under investigation (PUI) who were admitted to the PKRC on the date. PUI are individuals who have symptoms or exposure history suggestive of COVID-19, but have not been confirmed by laboratory testing.
- `admitted_covid`: The number of confirmed COVID-19 cases who were admitted to the PKRC on the date.
- `admitted_total`: The total number of admissions to the PKRC on the date. This is the sum of `admitted_pui` and `admitted_covid`.
- `discharged_pui`: The number of PUI who were discharged from the PKRC on the date. Discharge criteria may vary depending on the clinical and epidemiological situation.
- `discharged_covid`: The number of confirmed COVID-19 cases who were discharged from the PKRC on the date. Discharge criteria may vary depending on the clinical and epidemiological situation.
- `discharged_total`: The total number of discharges from the PKRC on the date. This is the sum of `discharged_pui` and `discharged_covid`.
- `pkrc_covid`: The number of confirmed COVID-19 cases who remained in the PKRC as of the date.
- `pkrc_pui`: The number of PUI who remained in the PKRC as of the date.
- `pkrc_total`: The total number of patients who remained in the PKRC as of the date. This is the sum of `pkrc_covid` and `pkrc_pui`.

### **d. Unique features or capabilities of Dabl:**
Dabl, the Data Analysis Baseline Library, brings unique features to exploratory data analysis (EDA). It automates visualizations, handling diverse data types intelligently, suggesting missing value strategies, and offering smart feature selection. With quick insights, concise statistical summaries, and basic model building capabilities, Dabl simplifies EDA. It supports preprocessing suggestions, handles mixed data types, and boasts a user-friendly interface. Dabl goes beyond by analyzing text data and incorporating cross-validation in model building, making it a swift and versatile tool for efficient data exploration.
