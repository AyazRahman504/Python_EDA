## 4. Implementation of Automated EDA Tools:

> ### 4.1 Load the dataset into the tool



```python
df = pd.read_csv('https://raw.githubusercontent.com/MoH-Malaysia/covid19-public/main/epidemic/hospital.csv')
```

> ### 4.2 Generate basic statistics and visualizations

```python
report = sv.analyze(df) #generate SweetViz report
```

> ### 4.3 Explore the relationships and pattern in the data

*   Once the report have been created, simply pass it into the 'show' function.

*   Open the SweetViz html report to explore the relationships and patterns.


```python
report.show_html('SweetViz_Report.html') #show report
```

<br>

*   Explore the correlations.




```python
report_correlation = sv.analyze(df, pairwise_analysis='on')
```

```python
report_correlation.show_html('Correlation_Report.html')
```

> ### 4.4 Unique features and capabilities


*   SweetViz can create visually appealing and interactive visualizations that provide an in-depth overview of the dataset. It may generate a variety of visualizations, including correlation matrices, distribution plots, and scatter plots.

*   It can includes the summary information about the type, unique values, missing values, duplicate rows, most frequent values and many more.

*   It also consist numeric analysis such as min/max/range, quartiles, mean, mode, standard deviation, sum, median absolute deviation, coefficient of variation, kurtosis and skewness.

*   SweetViz generates EDA reports as interactive HTML files. This makes it simple to share and explore the insights with others, even if they do not have the SweetViz library.

## 5. Pros and Cons Analysis:

> SweetViz provides a user-friendly and accessible solution for exploratory data analysis (EDA) in Google Colab. The tool's ease of use is essential, as it requires little coding to provide comprehensive data reports. Its comprehensiveness shows through, offering a full viewpoint of both numerical and categorical variables in the Malaysian hospital dataset. Furthermore, SweetViz's high-density visualizations were useful in understanding the distributions of crucial variables such hospital beds, admissions, and discharges. The reports are also provided as interactive HTML files, allowing for simple sharing and collaboration.

> SweetViz excels in visualization, yet it has significant limits. It is primarily intended for usage with Pandas Dataframes, and further conversion processes may be required if the data is not in correct format. Furthermore, as SweetViz is more focused on visualizations than on doing in-depth statistical tests, users who require extended statistical studies may need to supplement SweetViz with other tools. While customization options for analyses are available, the level of modification for the design of HTML reports may be more limited when compared to other tools.
