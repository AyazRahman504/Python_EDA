# Exploratory Data Analysis (EDA) in Python using Sweetviz

```python
pip install sweetviz
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/drshahizan/Python_EDA/main/assignment/bdm/Ayam%20Rendang/mudah-apartment-kl-selangor.csv')

# Separate X and Y
X = df.drop('prop_name', axis=1)
Y = df['prop_name']
```

```python
X
```

# Analyze The Dataset

```python
df
```

```python
import sweetviz as sv
analyze_report = sv.analyze(df)
analyze_report.show_html('analyze.html', open_browser=False)
```

```python
import IPython
IPython.display.HTML('analyze.html')
```

# Side by Side Comparison of Train vs Test Set

## Data Split using 80/20 Split Ratio

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train
```

```python
X_test
```

## Comparison Report

```python
import sweetviz as sv
compare_report = sv.compare([X_train, 'Train'], [X_test, 'Test'])
compare_report.show_html('compare.html', open_browser=False)
```

```python
import IPython
IPython.display.HTML('compare.html')
```
