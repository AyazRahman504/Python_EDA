# **Feature Preprocessing**

In the feature processing stage, we intend to address missing values by removing rows that contain any missing values in the dataset.

```python
# Drop rows with missing values in specific columns
columns_with_missing_values = ['bed', 'bath', 'acre_lot', 'city', 'house_size', 'price']

# Drop rows where any of the specified columns have missing values
data_cleaned = data.dropna(subset=columns_with_missing_values)
```

```python
# Display the cleaned DataFrame
print("Cleaned data:")
print(data_cleaned)
```


After eliminating the null values, the dataset now comprises 413,147 remaining rows.

```python
# Verify that there are no missing values after cleaning
print("Missing values after cleaning:")
print(data_cleaned.isnull().sum())
```

there are no more missing values in the dataset. The dataset is prepared to perform the next step of analysis

# **Feature Transformation**  

In this section, we will convert the values in the "house_size" column from square feet (sqft) to acres using the conversion formula: divide each value by 43,560, as 1 acre is equivalent to 43,560 square feet.

```python
# Converting house_size(sqrt) to acre
# 1 acre = 43560 square feet

data_cleaned['house_size'] = round(data_cleaned['house_size'] / 43560, 2)
data_cleaned['house_size']
```

```python
data_cleaned.head
```

# **Feature Creation**

Now, we will introduce a new column named "backyard" by executing arithmetic operations between the "acre_lot" and "house_size" columns. Specifically, the "backyard" column is derived as the difference between "acre_lot" and "house_size."

```python
# Create an initial 'backyard' column filled with placeholder values (e.g., 0)
data_cleaned['backyard'] = 0

#Arithmetic operations to calculate values of 'backyard'
data_cleaned['backyard'] = data_cleaned['acre_lot'] - data_cleaned['house_size']
```

```python
data_cleaned.head
```

```python
data_cleaned = data_cleaned.loc[data_cleaned['backyard'] > 0]
```

```python
# Assuming 'column_name' is the name of the column you want to check for negative values
# Replace 'column_name' with the actual name of your column

if (data_cleaned['backyard'] < 0).any():
    print("There are negative values in the column.")
else:
    print("There are no negative values in the column.")
```

# **Visualization**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'bed' and 'bath' are columns in your DataFrame
plt.figure(figsize=(12, 6))

# Histogram for 'bed' counts
plt.subplot(1, 2, 1)
sns.histplot(data_cleaned['bed'], bins=20, kde=True, color='green')
plt.title('Distribution of Bed Counts')
plt.xlabel('Number of Beds')
plt.ylabel('Frequency')

# Histogram for 'bath' counts
plt.subplot(1, 2, 2)
sns.histplot(data_cleaned['bath'], bins=20, kde=True, color='purple')
plt.title('Distribution of Bath Counts')
plt.xlabel('Number of Baths')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()





```

BAR PLOT BETWEEN AVERAGE PRICE AND STATES

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar Plot of Average Price by State
plt.figure(figsize=(12, 6))
sns.barplot(x='state', y='price', data=data_cleaned, estimator=np.mean)
plt.title('Average Price by State')
plt.xlabel('State')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'bed', 'bath', and 'state' are columns in your DataFrame
# Group by 'state' and calculate the average 'bed' and 'bath' counts
average_bed_bath_state = data_cleaned.groupby('state')[['bed', 'bath']].mean().reset_index()

# Melt the DataFrame to long format for plotting
melted_data = pd.melt(average_bed_bath_state, id_vars='state', var_name='Variable', value_name='Count')

# Create a grouped bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='state', y='Count', hue='Variable', data=melted_data, palette='viridis')
plt.title('Average Bed and Bath Counts Across States')
plt.xlabel('State')
plt.ylabel('Average Count')
plt.xticks(rotation=45)
plt.legend(title='Variable')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'backyard' and 'price' are columns in your DataFrame
backyard = data_cleaned['backyard']
price = data_cleaned['price']

# Set the range of backyard sizes you want to focus on
min_backyard_size = 0
max_backyard_size = 10

# Filter the data for the specified range of backyard sizes
filtered_data = data_cleaned[(backyard >= min_backyard_size) & (backyard <= max_backyard_size)]

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='backyard', y='price', data=filtered_data, scatter_kws={'s': 50, 'alpha': 0.5}, line_kws={'color': 'red', 'linewidth': 2})
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Backyard Size (acres)')
plt.ylabel('Price')
plt.grid(True)
plt.show()





```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'backyard' and 'state' are columns in your DataFrame
# Calculate the average backyard size for each state
average_backyard_size_state = data_cleaned.groupby('state')['backyard'].mean().reset_index()

# Create a point plot
plt.figure(figsize=(12, 6))
sns.pointplot(x='state', y='backyard', data=average_backyard_size_state, color='green', ci=None)
plt.title('Average Backyard Size Across States')
plt.xlabel('State')
plt.ylabel('Average Backyard Size')
plt.xticks(rotation=45)
plt.show()



```
