#About Dataset
**Job Dataset**

This dataset provides a comprehensive collection of synthetic job postings to facilitate research and analysis in the field of job market trends, natural language processing (NLP), and machine learning. Created for educational and research purposes, this dataset offers a diverse set of job listings across various industries and job types.

-------------------------------------------------------------------------

**Data Dictionary**

`Job Id` : A unique identifier for each job posting.

`Experience` : The required or preferred years of experience for the job.

`Qualifications` : The educational qualifications needed for the job.

`Salary Range` : The range of salaries or compensation offered for the position.

`Location` : The city or area where the job is located.

`Country` : The country where the job is located.

`Latitude` : The latitude coordinate of the job location.

`Longitude` : The longitude coordinate of the job location.

`Work Type` : The type of employment (e.g., full-time, part-time, contract).

`Company Size` : The approximate size or scale of the hiring company.

`Job Posting Date` : The date when the job posting was made public.

`Preference` : Special preferences or requirements for applicants (e.g., Only Male or Only Female, or Both)

`Contact Person` : The name of the contact person or recruiter for the job.

`Contact` : Contact information for job inquiries.

`Job Title` : The job title or position being advertised.

`Role` : The role or category of the job (e.g., software developer, marketing manager).

`Job Portal` : The platform or website where the job was posted.

`Job Description` : A detailed description of the job responsibilities and requirements.

`Benefits` : Information about benefits offered with the job (e.g., health
insurance, retirement plans).

`Skills` : The skills or qualifications required for the job.

`Responsibilities` : Specific responsibilities and duties associated with the job.

`Company Name` : The name of the hiring company.

`Company Profile` : A brief overview of the company's background and mission.

```python
df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'])
```

Split `Salary Range` into lower & upper bound

> First, we clean up the `Salary Range`

```python
# Remove the Letter K
df["Salary Range"] = df["Salary Range"].replace("K", "",regex=True)

# Remove the dollar '$' sign
df["Salary Range"] = df["Salary Range"].replace("$", "",regex=True)

# Remove the '-' sign
df["Salary Range"] = df["Salary Range"].replace("-", " ",regex=True)
```

> Split `Salary Range` into upper and lower bound

```python
# Remove the whitespaces
df["Salary Range"] = df["Salary Range"].apply(lambda x: x.strip(' '))

# split the Salary Estimate into upper and lower bound
df[['Salary Range Lower Bound (in K)','Salary Range Upper Bound (in K)']] = df["Salary Range"].str.split(" ",expand=True)

# drop the salary estimate column
df = df.drop(columns = "Salary Range")
```

```python
# Remove the dollar '$' sign from lower bound
df["Salary Range Lower Bound (in K)"] = df["Salary Range Lower Bound (in K)"].replace({'\$':""},regex=True)

# Remove the dollar '$' sign from upper bound
df["Salary Range Upper Bound (in K)"] = df["Salary Range Upper Bound (in K)"].replace({'\$':""},regex=True)
```

```python
df.head()
```

```python
# Convert the column from string to integer
df["Salary Range Lower Bound (in K)"] = df["Salary Range Lower Bound (in K)"].astype(int)

# Convert the column from string to integer
df["Salary Range Upper Bound (in K)"] = df["Salary Range Upper Bound (in K)"].astype(int)
```

## Feature Engineering with Automated Tools

###Automated Tool 1 - Featuretools


```python
!pip install -U featuretools
```

```python
import featuretools as ft

import warnings
warnings.filterwarnings('ignore')
```

```python
es = ft.EntitySet(id='job_data')
```

```python
# Creating a new DataFrame (job_titles_df) containing unique values from the "Job Title" column of the original DataFrame (df).
# The new DataFrame has a single column named "Job_Title"

job_titles_df = pd.DataFrame(df['Job Title'].unique(), columns=["Job_Title"])
```

```python
job_titles_df
```

```python
# Create an entity from the dataframe.
es = es.add_dataframe(dataframe=df, dataframe_name='jobs', index='Job Id', time_index='Job Posting Date')
```

```python
es
```

```python
es = es.add_dataframe(dataframe=job_titles_df, dataframe_name='job_titles', index='Job_Title')
```

```python
es
```

```python
es.add_relationship("job_titles", "Job_Title", "jobs", "Job Title")
```

```python
# Perform deep feature synthesis without specifying primitives
features, feature_names = ft.dfs(entityset=es, target_dataframe_name='job_titles',  max_depth = 1)
```

```python
# Histogram for Salary Range Lower Bound

plt.figure(figsize=(10, 6))
features['MEAN(jobs.Salary Range Lower Bound (in K))'].hist(bins=20)
plt.title('Distribution of Mean Salary Range Lower Bound')
plt.xlabel('Mean Salary Range Lower Bound (in K)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for salary range by work type
plt.figure(figsize=(12, 8))
sns.boxplot(x='MODE(jobs.Work Type)', y='MEAN(jobs.Salary Range Lower Bound (in K))', data=features)
plt.title('Box Plot of Mean Salary Range by Work Type')
plt.xlabel('Work Type')
plt.ylabel('Mean Salary Range Lower Bound (in K)')
plt.show()
```

```python
import matplotlib.pyplot as plt

top_jobs = features.nlargest(10, 'COUNT(jobs)')

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_jobs['COUNT(jobs)'], labels=top_jobs['MODE(jobs.Role)'], autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Job Roles by Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter dataframe for work type = 'intern'
intern_jobs = features[features['MODE(jobs.Work Type)'] == 'Intern']

# Get the top 10 intern jobs by count
top_intern_jobs = intern_jobs.nlargest(10, 'COUNT(jobs)')

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_intern_jobs['COUNT(jobs)'], labels=top_intern_jobs['MODE(jobs.Role)'], autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Intern Job Roles by Count')
plt.show()
```

Feature Engineering with Automated Tools

1) Creating an empty EntitySet with the identifier 'flights'
* `es = ft.EntitySet(id='job_data')`

2) Creating a new DataFrame (job_titles_df) containing unique values from the "Job Title" column of the original DataFrame (df). The new DataFrame has a single column named "Job_Title".
* `job_titles_df = pd.DataFrame(df['Job Title'].unique(), columns=["Job_Title"])`

3) Creating an entity from the df dataframe and add it into the empty EntitySet. This step is to make a foundation for the DFS by Featuretools that we will be performing later.
* `es = es.add_dataframe(dataframe=df, dataframe_name='jobs', index='Job Id', time_index='Job Posting Date')`

4) Establishing a relationship between the "jobs" entity and the "job_title" entity in the EntitySet. This step is very important because without a relationship the Featuretools cannot perform Deep Feature Synthesis.
* `es.add_relationship("job_titles", "Job_Title", "jobs", "Job Title")`

5) Performing the "Deep Feature Synthesis" function by using Featuretools on the entity set that we had created. In this step, we do not specify any primitives while performing the DFS therefore it could generate as much features as possible.
* `features, feature_names = ft.dfs(entityset=es, target_dataframe_name='job_titles',  max_depth = 1)`

6) By performing the tasks above, we had successfully applying Featuretools Automated Feature Engineering on our dataset as there is 101 new features generated. The main reason that we choose Featuretools as the Automated Feature Engineering Tool due to its ability to handle complex relationships, scalability, consistent framework, integration with EntitySets, time-aware feature engineering, a comprehensive library of primitives, compatibility with other libraries, and support from an active open-source community. It facilitates rapid prototyping and ensures consistency in feature engineering across datasets.
