#### **Featuretools:**
Featuretools is an open-source library which is used to carry out automated feature engineering. It is an excellent tool that accelerates the feature creation process, freeing up time for focusing on other facets of developing machine learning models. Stated differently, it renders your data "machine learning ready."


**Instaling and Importing Featuretools**

```python
!pip install featuretools
```

```python
import featuretools as ft
```

**Displya Data**

```python
spotify.head()
```

##### **Creating An EntitySet**

First, we initialize an EntitySet. We can also give the EntitySet a name by providing an id to the constructor.

```python
es = ft.EntitySet(id="spotify_data")
```

##### **Adding dataframes**

To start our feature engineering, we must first add dataframe to the EntitySet. Here, we will add two dataframe from spotify_data, which are tracks and artist.

```python
es = es.add_dataframe(
    dataframe_name="track",
    dataframe=spotify,
    index="track_id",
    time_index="year",
)

es
```

```python
es = es.add_dataframe(
    dataframe_name="artist",
    dataframe=spotify,
    index='artist_id',

)
es
```

##### **Run Deep Feature Synthesis(DFS)**

After adding the dataframes, we will proceed with run deep feature synthesis or dfs to automatically generate new features based on the relationship and pattern in the dataset.

```python
# Run deep feature synthesis with transformation primitives
feature_matrix, feature_defs = ft.dfs(entityset=es, target_dataframe_name='track',
                                      trans_primitives=['add_numeric', 'multiply_numeric'])

# Display the resulting feature matrix
print(feature_matrix.head())
```

The output above are the feature matrix that are automatically generated by Featuretools through dfs. Each row represent the unique value of the track, while the columns represent the various newly created features that are generated by combining or transforming the original columns.

Some insight from the output:

*   The current columns were combined and transformed to create a range of new features that are displayed in the feature matrix. These features produced based on the pattern found in the data
*   Transformation primitives like 'add_numeric' and'multiply_numeric' are used to generate the features. By merging or modifying the current features, Featuretools is able to generate new ones.

*   Certain features show how various original features interacted with one another. For instance, the product of "popularity" and "speechiness" is represented by the formula "popularity * speechiness."

3. Perform Train and Test Set

```python
X = spotify[['loudness']]
y = spotify[['mode']]
```

1. X = spotify[['loudness']]:

* spotify[['loudness']]: This part selects the column 'loudness' from the DataFrame spotify.
* The double square brackets ([['loudness']]) are used to create a DataFrame with a single column, rather than a Series.
* So, X is a DataFrame containing only the 'loudness' column from spotify.
2. y = spotify[['mode']]:

* spotify[['mode']]: This part selects the column 'mode' from the DataFrame spotify.
* Similar to the explanation for X, the double square brackets ([['mode']]) create a DataFrame with a single column.
* So, y is a DataFrame containing only the 'mode' column from spotify.

In summary, after executing these lines, X will be a DataFrame containing the 'loudness' column, and y will be a DataFrame containing the 'mode' column. This is a common setup in machine learning, where X represents the features (independent variables), and y represents the target variable (dependent variable) that you want to predict.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.43, random_state=52)
```

* X: The feature matrix.
* y: The target variable.
* test_size: The proportion of the dataset to include in the test split. In this case, it's set to 0.43, meaning 43% of the data will be used for testing, and the remaining 57% will be used for training.
* random_state: Seed for the random number generator. Setting this ensures reproducibility. If you use the same random_state value in the future, you'll get the same split.

X_train, X_test, y_train, y_test: These variables will hold the result of the data split.

* X_train: The subset of the feature matrix X that will be used for training.
* X_test: The subset of the feature matrix X that will be used for testing.
* y_train: The subset of the target variable y that corresponds to the training data.
* y_test: The subset of the target variable y that corresponds to the testing data.

After running this line of code, you can use X_train and y_train to train your machine learning model and X_test and y_test to evaluate its performance on unseen data. The specific proportions and random state values can be adjusted based on your requirements.

```python
model = AutoFeatRegressor()
model
```




* AutoFeatRegressor: This is a class provided by the autofeat library, and it stands for "Automatic Feature Regressor." The purpose of this class is to automatically engineer new features from the existing ones to improve the performance of regression models.

* model = AutoFeatRegressor(): This line creates an instance of the AutoFeatRegressor class and assigns it to the variable model. This instance represents the regression model with automatic feature engineering capabilities.

```python
X_train_feature_creation = model.fit_transform(X_train.to_numpy(), y_train.to_numpy().flatten())
```

* model.fit_transform(): This method is part of the AutoFeatRegressor class. The fit_transform method in scikit-learn is commonly used for both fitting a model and transforming the input data. In the context of AutoFeatRegressor, it not only fits the model but also applies automatic feature engineering to the input data.

* X_train.to_numpy(): Converts the training feature matrix X_train to a NumPy array. The fit_transform method typically expects the input data in the form of NumPy arrays.

* y_train.to_numpy().flatten(): Converts the training target variable y_train to a NumPy array and then flattens it into a 1-dimensional array. The flattening is done to ensure that y_train is a 1D array, as expected by the fit_transform method.

* X_train_feature_creation: This variable will hold the result of applying feature engineering to the training features. After executing this line, X_train_feature_creation will contain the original features along with the newly created features by the AutoFeatRegressor.

The purpose of using fit_transform is to simultaneously fit the AutoFeatRegressor model to the training data (X_train, y_train) and transform the features using the automatic feature engineering capabilities provided by the AutoFeatRegressor. The transformed features are stored in the variable X_train_feature_creation, and these features can be used for training a machine learning model.

```python
X_test_feature_creation = model.transform(X_test.to_numpy())
```

* model.transform(): This method is part of the AutoFeatRegressor class. Unlike fit_transform, transform is used only for applying previously learned transformations to new data. In this case, it applies the feature engineering transformations that were learned during the training phase to the test data.

* X_test.to_numpy(): Converts the test feature matrix X_test to a NumPy array. The transform method typically expects the input data in the form of NumPy arrays.

* X_test_feature_creation: This variable will hold the result of applying the previously learned feature engineering transformations to the test features. After executing this line, X_test_feature_creation will contain the transformed features that were created during the training phase using the AutoFeatRegressor.

The purpose of using transform on the test data is to ensure that the same feature engineering steps applied to the training data are consistently applied to the test data. This is important for maintaining the consistency of the data preprocessing pipeline between training and testing.
